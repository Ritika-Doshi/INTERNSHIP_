## Q8: Write a python program to display list of 50 Most expensive cars in the world.

```python
!pip install selenium
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.motor1.com/")
```

```python
search = driver.find_element(By.XPATH,"/html/body/div[10]/div[2]/div/div/div[3]/div/div/div/form/input")
search.send_keys('50 most expensive cars')
```

```python
trending= driver.find_element(By.XPATH,'/html/body/div[10]/div[10]/div[2]/div[5]/a[2]')
trending.click()
```

```python
Name=[]
Price=[]
```

```python
name_tags=driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in name_tags:
    name=i.text
    Name.append(name)
Name=Name[:50]
prices_tag=driver.find_elements(By.XPATH,'//p[strong]')
for i in prices_tag:
    price=i.text
    Price.append(price)
```

```python
df=pd.DataFrame({"Car name":Name,"Price":Price})
df
```
